```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('MoreIncomeData.csv')
data = data[2:53].reset_index()
data.head()
```

```python
# print the state names to a csv file, to be used for later analysis
states = data.State
states.to_csv('StateNames.csv')
```

```python
# Function to remove nondigits from values in a column
def remove_nondigits(col):
    values = []
    for val in col:
        val = str(val)
        newval = re.sub(r'\D', "", val)
        values.append(newval)
    return pd.Series(values)
```

```python
data = data.set_index('State')
data.columns = remove_nondigits(data)
data = data.reset_index()

data.head()
```

```python
cols = ['State', '2007', '2008', '200936', '201037', '2011', '2012', '201338', '2014', '2015', '2016', '2017']
```

```python
df = data[cols]
df.columns = ['State', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
df.head()
```

```python
stacked_data = df.set_index('State').stack().reset_index()
stacked_data.columns = ['Area', 'Year', 'MedianIncome']
stacked_data.MedianIncome = remove_nondigits(stacked_data.MedianIncome).astype(int)
stacked_data.head()
```

```python
stacked_data.to_csv('StateIncomeData.csv')
```
