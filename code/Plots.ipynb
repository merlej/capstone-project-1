```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Read in the U.S. counties data for the map
counties_data = gpd.read_file('UScounties_data.geojson')
counties_data.head()
```

```python
# Read in the unemployment data
# unemployment_data = pd.read_csv('CleanDataUnemployment.csv')
unemployment_data = pd.read_csv('UnemploymentData.csv')
unemployment_data = unemployment_data.set_index('State').drop(['US', 'PR'], axis=0).reset_index()
unemployment_data.head()
```

```python
unemployment_data = unemployment_data[['Area_name', 'Unemployment_rate_2017']]
unemployment_data.info()
```

```python
# Drop data for Peurto Rico, Alaska, and Hawaii
counties_data = counties_data[(counties_data.STATE!='72') & (counties_data.STATE!='02') & (counties_data.STATE!='15')]
                            

counties_data.info()
```
